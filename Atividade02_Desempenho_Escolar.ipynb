```python
import pandas as pd

# Carregar dataset
df = pd.read_csv("student-mat.csv", sep=";")

# Criar variÃ¡vel alvo: aprovado = G3 >= 10
df["aprovado"] = (df["G3"] >= 10).astype(int)
df.head()
```

```python
# ðŸ“Š AnÃ¡lise Descritiva
print(df.describe())
print("\nMÃ©dia G3:", df["G3"].mean())
print("Mediana G3:", df["G3"].median())
```

```python
# ðŸ”§ PrÃ©-processamento
# Separar X e y
X = df.drop(["G3", "aprovado"], axis=1)
y = df["aprovado"]

# One-hot encoding para variÃ¡veis categÃ³ricas
X = pd.get_dummies(X, drop_first=True)

# NormalizaÃ§Ã£o
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Modelos
dt = DecisionTreeClassifier(random_state=42)
svc = LinearSVC(max_iter=10000, random_state=42)

dt.fit(X_train, y_train)
svc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("ðŸŽ¯ Decision Tree Accuracy:", accuracy_score(y_test, dt.predict(X_test)))
print(confusion_matrix(y_test, dt.predict(X_test)))
print(classification_report(y_test, dt.predict(X_test)))

print("ðŸŽ¯ Linear SVC Accuracy:", accuracy_score(y_test, svc.predict(X_test)))
print(confusion_matrix(y_test, svc.predict(X_test)))
print(classification_report(y_test, svc.predict(X_test)))
```
